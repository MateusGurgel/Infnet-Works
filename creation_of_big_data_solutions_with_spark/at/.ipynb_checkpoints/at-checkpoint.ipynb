# Engenharia de Atributos Avan√ßada

Crie novas vari√°veis que possam melhorar a previs√£o, por exemplo:

    faixa_etaria (categorias por idade)
    cidade_grande (bin√°rio para SP e RJ)
    gastos_por_idade (raz√£o entre gastos e idade)
    intera√ß√µes entre vari√°veis (ex: escolaridade √ó cidade)

Justifique brevemente a cria√ß√£o de cada nova vari√°vel.

Nessa pipeline, vamos criar:

**Faixa et√°ria:** Classificadores de regress√£o log√≠stica n√£o conseguem agrupar idades, al√©m disso, classificadores de √°rvores, como GBoost e Decision trees tamb√©m se beneficiam de tal feature.

**cidade_grande:** Os classificadores n√£o s√£o capazes de distinguir diferen√ßas entre cidades pequenas ou grandes, tentar apontar, pode levar algoritmos de √°rvores a decis√µes mais limpas, necessitando de uma quantidade menor de √°rvores para classificar cada entrada.

**gastos_por_idade:** Os modelos de √°rvore n√£o conseguem realizar contas matem√°ticas, ent√£o adicionar essa ratio pode melhorar a performance.

**pai_de_familia:** Essa √© uma agrega√ß√£o entre estado_civil == true & g√™nero == homem & idade > 27. Pode indicar se o homem provavelmente √© um pai de familia. 27 anos √© a idade m√©dia em que os pais tem seus filhos, segundo o IBGE.





```python
from pyspark.sql.functions import when, col, round as spark_round

df_client_with_features = df_client.withColumn("faixa_etaria",
    when(col("idade") <= 25, "Jovem")
    .when((col("idade") > 25) & (col("idade") <= 35), "Jovem_Adulto")
    .when((col("idade") > 35) & (col("idade") <= 50), "Adulto")
    .when((col("idade") > 50) & (col("idade") <= 65), "Maduro")
    .otherwise("Idoso")
)

df_client_with_features = df_client_with_features.withColumn("gastos_por_idade",
    spark_round(col("gastos_mensais") / col("idade"), 2)
)

df_client_with_features = df_client_with_features.withColumn("cidade_grande",
    when((col("cidade") == "S√£o Paulo") | (col("cidade") == "Rio de Janeiro")
         | (col("cidade") == "Salvador"), 1)
    .otherwise(0)
)


df_client_with_features.show()
```

‚öôÔ∏è 4. Pr√©-processamento Aprimorado

    Converta as vari√°veis categ√≥ricas em num√©ricas.
    Normalize os atributos num√©ricos se necess√°rio.
    Monte um vetor de atributos final com VectorAssembler.


Convertendo vari√°veis categ√≥ricas em num√©ricas

```python
from pyspark.ml.feature import OneHotEncoder, StringIndexer
from pyspark.ml import Pipeline

categorical_cols = ["genero", "cidade", "escolaridade", "cidade_grande", "faixa_etaria", "estado_civil"]
encoded_cols = []

stages = []
for col in categorical_cols:
    indexer = StringIndexer(inputCol=col, outputCol=f"{col}_index")
    stages.append(indexer)


    encoded_output = f"{col}_encoded"
    encoder = OneHotEncoder(inputCol=f"{col}_index", outputCol=encoded_output)
    stages.append(encoder)
    encoded_cols.append(encoded_output)
```

Normalizando os atributos num√©ricos, montando o vector final e gerando a pipeline

```python
from pyspark.ml.feature import StandardScaler

# Renda, como vari√°vel target n√£o foi inclu√≠da no assembler
numeric_cols = ["idade", "gastos_mensais", "gastos_por_idade"]

assembler = VectorAssembler(
    inputCols=numeric_cols,
    outputCol=f"numeric_vector"
)
stages.append(assembler)

scaler = StandardScaler(
    inputCol=f"numeric_vector",
    outputCol=f"numeric_vector_scaled",
    withStd=True,
    withMean=True
)

stages.append(scaler)

final_assembler = VectorAssembler(
    inputCols=[*encoded_cols, "numeric_vector_scaled"],
    outputCol="final_vector"
)

stages.append(final_assembler)

pipeline = Pipeline(stages=stages)
model = pipeline.fit(df_balanced_df)
df_final = model.transform(df_balanced_df)

df_final.show()
```

A decis√£o de normalizar todos os campos √© por conta da performance do meu computador.

Estou em um notebook com um i5, qualquer algoritmo que n√£o seja a regress√£o linear demora muito para ser treinado.

A regress√£o linear √© muito sens√≠vel a dados n√£o normalizados em escalas diferentes.

Al√©m disso, o Kmeans tamb√©m se beneficia da regulariza√ß√£o dos dados.

üîÑ 5. Ajuste de Hiperpar√¢metros

    Escolha e treine um modelo de regress√£o entre:
        Linear Regression
        Decision Tree Regressor
        Gradient-Boosted Trees Regressor
    Realize ajuste de hiperpar√¢metros com ParamGridBuilder.
    Use valida√ß√£o cruzada (CrossValidator) com no m√≠nimo 3 folds.


```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.regression import LinearRegression

train_data, test_data = df_final.randomSplit([0.8, 0.2], seed=42)

linear_regressor = LinearRegression(
    featuresCol="final_vector",
    labelCol="renda",
)

paramGrid = ParamGridBuilder() \
    .addGrid(linear_regressor.regParam, [0.01, 1.0, 5.0, 10.0]) \
    .addGrid(linear_regressor.elasticNetParam, [0.0, 0.1, 0.5, 0.8, 1.0]) \
    .addGrid(linear_regressor.maxIter, [25, 50, 100, 200]) \
    .build()

evaluator = RegressionEvaluator(
    labelCol="renda",
    predictionCol="prediction",
    metricName="rmse"
)

cross_val = CrossValidator(
    estimator=linear_regressor,
    estimatorParamMaps=paramGrid,
    evaluator=evaluator,
    numFolds=5,
    seed=42
)

cv_model = cross_val.fit(train_data)
best_model = cv_model.bestModel

print("Melhores Parametros de Treino:")
print(f"Regularization Parameter (regParam): {best_model.getRegParam()}")
print(f"Elastic Net Parameter (elasticNetParam): {best_model.getElasticNetParam()}")
print(f"Max Iterations (maxIter): {best_model.getMaxIter()}")
print(f"Standardization: {best_model.getStandardization()}")
```

üìà 6. Avalia√ß√£o do Modelo

    Utilize m√©tricas como RMSE e R¬≤ para medir desempenho.
    Compare diferentes configura√ß√µes e justifique a melhor.


```python
test_predictions = cv_model.bestModel.transform(test_data)

rmse_evaluator = RegressionEvaluator(labelCol="renda", predictionCol="prediction", metricName="rmse")
mae_evaluator = RegressionEvaluator(labelCol="renda", predictionCol="prediction", metricName="mae")
r2_evaluator = RegressionEvaluator(labelCol="renda", predictionCol="prediction", metricName="r2")

rmse = rmse_evaluator.evaluate(test_predictions)
mae = mae_evaluator.evaluate(test_predictions)
r2 = r2_evaluator.evaluate(test_predictions)

print(f"RMSE (Root Mean Square Error): R$ {rmse:.2f}")
print(f"MAE (Mean Absolute Error): R$ {mae:.2f}")
print(f"R¬≤ (Coeficiente de Determina√ß√£o): {r2:.4f}")
```

```python
from pyspark.ml import PipelineModel

loaded_pipeline = PipelineModel.load("./pipeline")

df_final = loaded_pipeline.transform(df_balanced_df)
```

Treine um modelo com KMeans, variando k de 3 a 5 e utilizando silhouette_score para avaliar o modelo.

```python
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml.clustering import KMeans

k_values = [2, 3, 4, 5, 6, 8]
silhouette_scores = []
models = {}
predictions_k = {}

for k in k_values:
    kmeans = KMeans(
        featuresCol="final_vector",
        predictionCol=f"cluster_id",
        k=k,
        seed=42,
        maxIter=100
    )

    model = kmeans.fit(df_final)
    models[k] = model

    predictions = model.transform(df_final)
    predictions_k[k] = predictions

    evaluator = ClusteringEvaluator(
        featuresCol="final_vector",
        predictionCol=f"cluster_id",
        metricName="silhouette"
    )

    silhouette_score = evaluator.evaluate(predictions)
    silhouette_scores.append(silhouette_score)

    print(f"k:{k} ,Silhouette Score: {silhouette_score:.4f}")

```

```python
from pyspark.sql.functions import col

total_k_analysed = [2, 3, 6]

for total_k in total_k_analysed:
    print(f"An√°lise da clusteriza√ß√£o com : {total_k} K")
    prediction = predictions_k[total_k]
    for k in range(0, total_k):

        print("_" * 50)
        print(f"Cluster {k + 1}")
        k_prediction = prediction.filter(col(f"cluster_id") == k)

        # M√©dia de idade
        k_prediction.agg(
            avg("idade").alias("idade_media")
        ).show()

        # M√©dia de renda
        k_prediction.agg(
            avg("renda").alias("renda_m√©dia")
        ).show()

        k_prediction.groupBy("faixa_etaria").count().orderBy(col("count").desc()).show()
        k_prediction.groupBy("cidade").count().orderBy(col("count").desc()).show()
        k_prediction.groupBy("estado_civil").count().orderBy(col("count").desc()).show()
        k_prediction.groupBy("escolaridade").count().orderBy(col("count").desc()).show()
        k_prediction.groupBy("cidade").count().orderBy(col("count").desc()).show()
        k_prediction.groupBy("genero").count().orderBy(col("count").desc()).show()
        k_prediction.groupBy("cidade_grande").count().orderBy(col("count").desc()).show()
```

Observando o comportamento dos clusters em k's diferentes, pude observar que:

-  Quanto menor o K, melhor os clusters s√£o consolidados.

-  As cidades parecem n√£o ter tanto impacto, por mais que podemos observar tend√™ncias, onde cidades maiores aparecem com mais frequ√™ncia em clusters com maior renda m√©dia.

-  O estado civil nos clusters mais pobres √© homog√™neo, enquanto nos mais ricos, os casais prevalecem.

-  Em clusters com a renda m√©dia maior, a escolaridade tamb√©m tende a ser maior.

-  O genero das pessoas √© homogeneo qunado o K √© pequeno, mas quando o K aumenta, passa a ser um fator importante para a clusteriza√ß√£o.
